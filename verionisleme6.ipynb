```python
import numpy as np 
import pandas as pd 

v1=np.array([1,3,6,np.nan,7,1,np.nan,9,15])
v2=np.array([7,np.nan,5,8,12,np.nan,np.nan,2,3])

v3=np.array([np.nan,10,11,8,np.nan,9,14,30,45])
             
df=pd.DataFrame({"v1":v1,"v2":v2,"v3":v3})
df
```

```python
#sayisaldeğer atama
```

```python
df["v1"].fillna(0)#eksik değere 0 
```

```python
df
```

```python
df.fillna(df["v1"].mean())
```

```python
#tüm değişkenler için birinci yol
df.apply(lambda x:x.fillna(x.mean()),axis=0)
```

```python
# ikinici yol
```

```python
df.fillna(df.mean()[:])
```

```python
df.fillna(df.mean()["v1":"v2"])
```

```python
df["v3"].fillna(df["v3"].median())
```

```python
# ücüncü yol 
```

```python
df.where(pd.notna(df),df.mean(),axis="columns")# tüm eksik deperlrin ortalama miktara
```
